```python
#converting into lowercase
def lower(df, col):
    for i in range(df[col].shape[0]):
        curr = df[col].values[i]
        df[col].values[i] = str(curr).lower()
    return df
```

```python
cat_cols = list(set(df.columns) -set(df.describe().columns))
cat_cols
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
#'5'.isdigit()

#removing stopwords and punctuations
punc = ['.', ',', '/', ':', ';', '!', '@', '#', '$', '%', '&', '*', '•', '\\', '?','(', ')', '-', " "]
for i in range(df.job_desc.shape[0]):
    doc = nlp(df.job_desc.values[i])
    val = " "
    for each in doc:
        if not(each.is_stop) and (str(each) not in punc) and (not(str(each).isdigit())):
            val = val + " " + str(each)
    df.job_desc.values[i] = val       
```

```python
#removing adj and verb
for i in range(df.job_desc.shape[0]):
    doc = nlp(df.job_desc.values[i])
    val = " "
    for each in doc:
        if each.pos_ != 'ADJ' and each.pos_ != 'VERB':
            val = val + " " + str(each)
    df.job_desc.values[i] = val       
```

```python
not_rqd=['b.sc', 'm.tech', 'b.e', 'etc', 'b.tech', 'role', 'location', 'months', 'year', 'knowledge', 'description', 'ug', 'pg']
for i in range(df.job_desc.shape[0]):
    doc = nlp(df.job_desc.values[i])
    val = " "
    for each in doc:
        if str(each) not in not_rqd:
            val = val + " " + str(each)
    df.job_desc.values[i] = val       
```
