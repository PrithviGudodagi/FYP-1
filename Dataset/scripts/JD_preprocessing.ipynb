```python
#Libraries
import pandas as pd
```

```python
#Loading the dataset
df = pd.read_csv("/home/nikita/College/FYP/Dataset/dataset/CSV/JD.csv", sep='delimiter', header=None)
print("Dataset loaded!")
```

```python
df.columns =['Description']
df
```

```python
#Data Preprocessing
#To remove number
df = df.replace('\d+', '', regex=True)
#Converting dataset into lowercase
df['Description'] = df['Description'].str.lower()
#To remove everything but letters
df = df.replace('[|/,:.+]','', regex=True)
#Convert dataset to string
df['Description']=df['Description'].apply(str)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]
```

```python
df['Description'] = df.apply(lemmatize_text)
```

```python
#Exporting
df.to_excel("/home/nikita/College/FYP/Dataset/dataset/JD_processed.xlsx", index = False, header=True)
# df.to_csv("/home/nikita/College/FYP/Dataset/dataset/syllabus_dataset_processed.csv", index = False, header=True)
```
